### 2.1. Etiquetas de variables/columnas: no usar caracteres especiales

```python
# Importo Modulo para detectar encoding
import chardet
```

```python
# Miramos los primeros diez mil bytes para adivinar la codificación de caracteres
with open("./datasets/Building_Permits_Checked.csv", 'rb') as rawdata:
    result = chardet.detect(rawdata.read(10000))

# Imprimio el enconding detectado.
print(result)
```

```python
SanFrancisco_Building_Permits.columns[SanFrancisco_Building_Permits.columns != SanFrancisco_Building_Permits.columns.str.extract(r'^(\w+)$')]
```
