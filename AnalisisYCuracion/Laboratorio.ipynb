### 2.2. Tratar valores faltantes

Veamos cuantos valores nulos tenemos

```python
# Obtengo el nro de datos faltantes por cada columna, listo solo las columnas que tienen faltantes.
missing_values_count = SanFrancisco_Building_Permits.isnull().sum()
missing_values_count[missing_values_count > 0]
```

```python
#Vemos que hay muchos registros con valores faltantes. Analizamos que pasa si borramos los registros con datos faltantes.
len(SanFrancisco_Building_Permits.dropna())
```

```python
# El efecto de hacer un dropna tiene una incidencia muy alta por lo que debemos tratar dichos valores.
# Podemos hacer un tratamiento para los campos numericos como los vinculados a street y poner un valor 0 y a los description string ponerles un vacio.
# Una feature a tomar con tratamiento distinto puede ser estimate_cost para la cual podemos rellenar con la media. Aplicamos dicho valor
SanFrancisco_Building_Permits.Estimated_Cost.describe()
```

```python
# reemplazamos NaN por la media.
Estimated_Cost_dna= SanFrancisco_Building_Permits.Estimated_Cost.fillna(SanFrancisco_Building_Permits.Estimated_Cost.mean)
```

```python
# Verificamos que no queden NaN
Estimated_Cost_dna.isnull().sum()
```

```python
SanFrancisco_Building_Permits.Estimated_Cost = Estimated_Cost_dna
```

### 2.3. Codificar variables

Las variables categóricas deben ser etiquetadas como variables numéricas, no como cadenas

```python
from sklearn import preprocessing
```

```python
column='Current_Status'
# Create a label (category) encoder object
le = preprocessing.LabelEncoder()
# Fit the encoder to the pandas column
le.fit(SanFrancisco_Building_Permits[column])
```

```python
# View encoder mapping
dict(zip(le.classes_,le.transform(le.classes_)))
```

```python
# Apply the fitted encoder to the pandas column
SanFrancisco_Building_Permits[column] = le.transform(SanFrancisco_Building_Permits[column]) 
SanFrancisco_Building_Permits.head()
```

```python
# Reversing encoding
SanFrancisco_Building_Permits[column] = le.inverse_transform(SanFrancisco_Building_Permits[column]) 
SanFrancisco_Building_Permits.head()
```
