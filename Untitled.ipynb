```python
# Pesos y altura
# ¿Quien gasta mas en apariencia? ¿la gente morruda o mas bien flaca?
# Para realizar este estudio dividiremos entre hombres y mujeres y calcularemos el indice de masa corporal de cada
# individuo como una nueva variable

dataset_mujeres = dataset[dataset["Gender"]=="female"]
dataset_hombres = dataset[dataset["Gender"]=="male"]

# El indice de masa corporal (IMC) depende de la altura

dataset_mujeres_na = dataset_mujeres.dropna()
dataset_hombres_na = dataset_hombres.dropna()

dataset_na = dataset.dropna()

dataset_mujeres_na["IMC"] = dataset_mujeres_na.apply(lambda row: row.Weight / ((row.Height/100)*(row.Height/100)), axis=1)
dataset_hombres_na["IMC"] = dataset_hombres_na.apply(lambda row: row.Weight / ((row.Height/100)*(row.Height/100)), axis=1)

dataset_na["IMC"] = dataset_na.apply(lambda row: row.Weight / ((row.Height/100)*(row.Height/100)), axis=1)

# Teniendo el indice de masa corporal se pueden establecer las siguientes categorias

#      IMC            |  Nivel de peso
#------------------------------------
# Por debajo de 18.5  |  Bajo peso
# 18.5 – 24.9         |  Normal
# 25.0 – 29.9         |  Sobrepeso
# 30.0 o más          |  Obeso

#procederemos a añadir una columna nueva al dataset con estas categorias para cada persona


def categoriza_peso(data):
    if data.IMC < 18.5:
        return "bajo_peso"
    if data.IMC >=18.5 and data.IMC < 25:
        return "normal"
    if data.IMC >= 25 and data.IMC < 30:
        return "sobrepeso"
    if data.IMC >= 30:
        return "obeso"
    
dataset_mujeres_na["nivel_de_peso"] = dataset_mujeres_na.apply(categoriza_peso, axis=1)

dataset_hombres_na["nivel_de_peso"] = dataset_hombres_na.apply(categoriza_peso, axis=1)


dataset_na["nivel_de_peso"] = dataset_na.apply(categoriza_peso, axis=1)

orden_peso = ["bajo_peso", "normal", "sobrepeso", "obeso"]

dataset_na= dataset_na.astype(dtype={"nivel_de_peso":"category"}, ordered=True, 
                       categories={"nivel_de_peso":orden_peso})

dataset_na.nivel_de_peso = dataset_na.nivel_de_peso.cat.codes
```
