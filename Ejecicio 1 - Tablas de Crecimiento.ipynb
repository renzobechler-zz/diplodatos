```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

### Percentiles de peso por edad (niños y niñas desde recien nacidos hasta 36 meses)

```python
pesos_por_edad = pd.read_csv("./datasets/wtageinf.csv")

```

```python
pd.set_option('display.max_columns',76) # Para poder visualizar todas las columnas

pesos_por_edad.describe()
```

```python
pesos_por_edad_ninios = pesos_por_edad[pesos_por_edad.Sex == 1]
pesos_por_edad_ninios = pesos_por_edad_ninios.sort_values(by="Agemos")

pesos_por_edad_ninias = pesos_por_edad[pesos_por_edad.Sex == 2]
pesos_por_edad_ninias = pesos_por_edad_ninias.sort_values(by="Agemos")

```

```python
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)

for column in pesos_por_edad_ninios.columns[5:]:
    ax.plot(pesos_por_edad_ninios.Agemos, pesos_por_edad_ninios[column],label=column);

plt.title("Pesos por edad [niños hasta 36 meses]")
plt.ylabel("peso [kg]")
plt.xlabel("edad [meses]")
plt.grid(True)
plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)

for column in pesos_por_edad_ninias.columns[5:]:
    ax.plot(pesos_por_edad_ninias.Agemos, pesos_por_edad_ninias[column],label=column);

plt.title("Pesos por edad [niñas hasta 36 meses]")
plt.ylabel("peso [kg]")
plt.xlabel("edad [meses]")
plt.grid(True)
plt.legend()
plt.show()
```

### Percentiles de altura por año (niños y niñas desde recien nacidos hasta 36 meses)

```python
altura_por_edad = pd.read_csv("./datasets/lenageinf.csv")
```

```python
altura_por_edad.describe()
```

```python
altura_por_edad_numeric = altura_por_edad
altura_por_edad_numeric = altura_por_edad_numeric.apply(pd.to_numeric, errors="coerce")
altura_por_edad_numeric.dtypes
```

```python
altura_por_edad_numeric.Sex.isnull()

altura_por_edad_ninios = altura_por_edad_numeric[altura_por_edad_numeric.Sex== 1.0]
altura_por_edad_ninios = altura_por_edad_ninios.sort_values(by="Agemos")

altura_por_edad_ninias = altura_por_edad_numeric[altura_por_edad_numeric.Sex == 2.0]
altura_por_edad_ninias = altura_por_edad_ninias.sort_values(by="Agemos")


```

```python
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)

for column in altura_por_edad_ninios.columns[5:14]:
    ax.plot(altura_por_edad_ninios.Agemos, altura_por_edad_ninios[column],label=column);

plt.title("Altura por edad [niños hasta 36 meses]")
plt.ylabel("altura [cm]")
plt.xlabel("edad [meses]")
plt.grid(False)
plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)

for column in altura_por_edad_ninias.columns[5:14]:
    ax.plot(altura_por_edad_ninias.Agemos, altura_por_edad_ninias[column],label=column);

plt.title("Altura por edad [niñas hasta 36 meses]")
plt.ylabel("altura [cm]")
plt.xlabel("edad [meses]")
plt.grid(True)
plt.legend()
plt.show()
```
